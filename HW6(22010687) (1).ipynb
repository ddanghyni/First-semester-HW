### ```선형회귀분석```

````python
beta0 ,beta1, sigma = 5, 2, 3
````

````python
import numpy as np

x = [0.49, 0.59, 1.16, 1.40, 2.06, 2.42, 4.16, 4.85, 5.14, 5.23, 5.96, 6.09, 6.84, 6.85, 7.45, 7.61, 7.88, 8.23, 9.22, 9.81]
x = np.array(x)
len(x)
````

## STEP 1

````python
e_i = np.random.normal(scale = sigma, size = (len(x)))
````

````python
Y_i = []
for x_i, e in zip(x, e_i):
    Y_i.append(beta0 + beta1 * x_i + e)
````

````python
import matplotlib.pyplot as plt
xs = np.linspace(0, 10, 100)
plt.scatter(x, Y_i)
plt.plot(xs, beta0 + beta1 * xs, 'g:')
plt.show()
````

## STEP 2

````python
def mean(v):
    return sum(v) / len(v)
````

````python
def S(v):
    return sum((t_i - mean(v)) ** 2 for t_i in v)


def Sxy(v, w):
    return sum((x_i - mean(v)) * (y_i - mean(w))for x_i, y_i in zip(v, w))


S_xx, S_yy = S(x), S(Y_i)
S_xy = Sxy(x, Y_i)
Beta1_hat = S_xy / S_xx
Beta0_hat = mean(Y_i) - Beta1_hat * mean(x)
print("S_xx:",S_xx)
print("S_yy:",S_yy)
print("S_xy:",S_xy)
print("^β1:", Beta1_hat)
print("^β0:",Beta0_hat)
````

````python
import matplotlib.pyplot as plt
print("Estimate of beta1 :",Beta1_hat)
print("Estimate of beta0 :",Beta0_hat)
xs = np.linspace(0, 10, 100)
plt.scatter(x, Y_i)
plt.plot(xs, beta0 + beta1 * xs, 'g:', label = "real")
plt.plot(xs, Beta0_hat + Beta1_hat * xs, color = "Blue", label = "fitted")
plt.legend(loc = "upper left")
plt.show()
````
